# This notebook calculates the percent changes in voter turnout in each ward.

```python
import pandas as pd
```

```python
df_04_08 = pd.read_csv("mil_cnty_04_08.csv", usecols = ["GEOID", "TURNOUT04", "TURNOUT08"])
df_04_08
```

```python
df_12_20 = pd.read_csv("mil_cnty_12_20.csv", usecols = ["GEOID", "PCTBLK", "PCTHIS", "PCTASI", "TURNOUT12", "TURNOUT16", "TURNOUT20"])
df_12_20
```

```python
df = df_04_08
df["TURNOUT12"] = df_12_20.TURNOUT12
df["TURNOUT16"] = df_12_20.TURNOUT16
df["TURNOUT20"] = df_12_20.TURNOUT20
df["PCTBLK"] = df_12_20.PCTBLK
df
```

```python
df["CHANGE0408"] = df.TURNOUT04 - df.TURNOUT08
df["CHANGE0812"] = df.TURNOUT08 - df.TURNOUT12
df["CHANGE1216"] = df.TURNOUT12 - df.TURNOUT16
df["CHANGE1620"] = df.TURNOUT16 - df.TURNOUT20
df
```

```python
df["CHANGE0416"] = df.TURNOUT04 - df.TURNOUT16
df["CHANGE0420"] = df.TURNOUT04 - df.TURNOUT20
df
```

```python
df.to_csv("turnout_change.csv", index = False)
```
